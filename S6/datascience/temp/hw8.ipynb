```python
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium import webdriver



city_codes = {
    "Taipei": "63",
    "New Taipei": "65",
    "Taoyuan": "66",
    "Hsinchu County": "67",
    "Miaoli County": "68",
    "Taichung": "10002",
    "Changhua County": "10004",
    "Nantou County": "10005",
    "Yunlin County": "10007",
    "Chiayi County": "10008",
    "Chiayi City": "10009",
    "Tainan": "10010",
    "Kaohsiung": "10013",
    "Pingtung County": "10014",
    "Yilan County": "10015",
    "Hualien County": "10016",
    "Taitung County": "10017",
    "Kinmen County": "10018",
    "Lienchiang County": "10020",
    "Penghu County": "10004A",
    "Keelung": "09007",
    "Hsinchu City": "09020"
}

def get_temperature(city):
    if city.lower() == "all":
        for city_name, city_code in city_codes.items():
            temperature = get_temperature(city_name)
            print(f"The temperature in {city_name} is {temperature} degrees Celsius.")
    else:
       
        city_code = city_codes.get(city)
        if city_code:
            
            service = Service("./chromedriver.exe")
            driver = webdriver.Chrome(service=service)
            url = f"https://www.cwb.gov.tw/V8/E/W/County/County.html?CID={city_code}"
            driver.get(url)
            temperature_span = WebDriverWait(driver, 10).until(
                EC.presence_of_all_elements_located((By.CLASS_NAME, "tem-C")))
            temperature = temperature_span[0].text
            driver.quit()
            print(f"The temperature in {city} is {temperature} degrees Celsius.")
        else:
            print("Invalid city name.")

city_name = input("Enter a city name: ")
print(city_name)
get_temperature(city_name)
```

```python
import time
import pandas as pd

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select

def print_data():
    target = driver.find_element(By.XPATH, '//*[@id="PC_Week_MOD"]/thead/tr/th[1]').text
    temperature = driver.find_element(By.XPATH, '//*[@id="PC_Week_MOD"]/tbody/tr[1]/td[1]/p/span[1]').text
    print(target)
    print(temperature)

# initial
city = input()
url = 'https://www.cwb.gov.tw/V8/E/W/County/County.html'
driver = webdriver.Chrome()
driver.get(url)

# search
select_element = driver.find_element(By.ID, 'CID')
select = Select(select_element)
if city == 'all':
    for idx in range(1, 23):
        select_element = driver.find_element(By.ID, 'CID')
        select = Select(select_element)
        select.select_by_index(idx)

        print_data()

        time.sleep(1)


else:
    select.select_by_visible_text(city)
    print_data()

time.sleep(1)
driver.quit()
```

```python
from selenium.webdriver.common.by import By
from selenium import webdriver
import time
from bs4 import BeautifulSoup as BF
from selenium.webdriver.chrome.service import Service

city = input()

driverPath = "chromedriver.exe"
s = Service(executable_path=driverPath)

# 加啟動配置
option = webdriver.ChromeOptions()
option.add_argument('disable-infobars')
# 開啟chrome瀏覽器
browser = webdriver.Chrome(options=option, service=s)
url = 'https://www.cwb.gov.tw/V8/E/W/County/County.html?CID=63'
browser.get(url)                    

if (city == 'all'):
    # 爬取所有city名字
    s_ = browser.page_source
    sp_ = BF(s_,'lxml')
    select = sp_.find('select', {'id': 'CID'})
    options = select.find_all('option')
    # print(options)
    num_options = len(options)
    # print(num_options)
    
    all_city = []
    for option_ in options:
        value = option_['value']
        label = option_.text.strip()
        if (label != 'Choose a City/County'):
            all_city.append(label)
    
    # 把爬取到的所有city都跑一次去爬取他的天氣並印出來
    num_options = num_options - 1
    for i in range(num_options):
        txtBox = browser.find_element(By.ID, 'CID')
        txtBox.send_keys(all_city[i])
        time.sleep(5)
        source = browser.page_source
        sp = BF(source,'lxml')
        temp_span = sp.select_one('.tem .tem-C.is-active')
        weather = temp_span.text.strip()
        print(all_city[i])
        print(weather)
    
else:
    # 選擇city
    txtBox = browser.find_element(By.ID, 'CID')
    txtBox.send_keys(city)
    time.sleep(1)
    # 爬取天氣
    source = browser.page_source
    sp = BF(source,'lxml')
    temp_span = sp.select_one('.tem .tem-C.is-active')
    weather = temp_span.text.strip()
    print(city)
    print(weather)
    
browser.quit()
```

```python
import requests

from bs4 import BeautifulSoup

url = 'https://rate.bot.com.tw/xrt?Lang=en-US'

html = requests.get(url)

soup = BeautifulSoup(html.content, "html.parser")

tokens = []

prices = []

for i in range(1, 20):

    token = soup.select_one('#ie11andabove > div > table > tbody > tr:nth-child(' +str(i)+') > td.currency.phone-small-font > div > div.hidden-phone.print_show').get_text()

    price = soup.select_one('#ie11andabove > div > table > tbody > tr:nth-child('+str(i)+') > td:nth-child(3)').get_text()

    tokens.append(token)

    prices.append(price)

    

utoken = input()

uprice = input()

for i in range (0, 19):

    if utoken in tokens[i]:

        print(float(uprice) / float(prices[i]))
```
