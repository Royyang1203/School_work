中文：

給一未定係數的多項式：

<img src="img/q1_1.png">

此問題中需要你計算上式多項式微分過後的值。微分的定義如下：

<img src="img/q1_2.png">

所有的輸入與輸出都會是整數，詳細如下：

Input:

程式碼將有數行輸入。

第一行為所要代入的x值。

接下來數行會包含數個整數 a0 ,a1, ... , an-1, an ,為多項式的係數，最後會輸入End表示輸入結束。

Output:

計算出將x代入給定多項式的導函數之值。

sample input 1:

7

2

-10

3

End

sample output 1:

18

sample input 2:

2

2

0

0

1

End

sample output 2:

24

```python
arr = []

a = input()
b = input()
while b != 'End':
    arr += [b]
    b = input()


res = 0
for i in range(len(arr)):
    k = len(arr) - i - 1
    g = int(arr[i]) * (k) * (int(a) ** (k - 1))
    res += g

print(int(res))
```
