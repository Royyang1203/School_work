# Question 3

```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd


response = requests.get("https://news.ycombinator.com/")
soup = BeautifulSoup(response.content, 'html.parser')

# Find all elements with class 'athing' and store them in title_list
title_list = soup.find_all(class_='athing')

# Find all elements with class 'subtext' and store them in score_list
score_list = soup.find_all(class_='subtext')

res = []

for i, j in zip(title_list, score_list):           
	
	# Find the text of the element with class 'titleline' within the current element in title_list
	title = i.find(class_='titleline').text
	
	# If the current element in score_list has an element with class 'score', extract its numerical value
	if j.find(class_='score') != None:
		score = re.findall(r'\d+', j.find(class_='score').text)[0]
	else:
		score = None

	# If the current element in title_list has a link, extract its 'href' attribute value
	if i.find(class_='titleline').find('a', href=True) != None:
		href = i.find(class_='titleline').find('a', href=True)['href']
	else:
		href = None

	# Append the extracted information as a list to the results list
	res += [[title, score, href]]


df = pd.DataFrame(res, columns=['title', 'score', 'href'])
df
```

# Question 4

```python
import requests
from bs4 import BeautifulSoup

def get_lotto_numbers(period):
    url = f"https://www.taiwanlottery.com.tw/lotto/lotto649/history.aspx"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")

    # Find the target period row
    all_table = soup.find(id='Lotto649Control_history_dlQuery')
    table = all_table.find_all('table')
    for row in table:
        cur_period = row.find_all('td')
        if int(cur_period[5].text) == int(period):
            numbers = [int(number.text) for number in cur_period[13:20]]
            return numbers
    return None

def check_prize(user_numbers, winning_numbers):
    matched_numbers = set(user_numbers).intersection(set(winning_numbers[:-1]))
    matched_count = len(matched_numbers)

    if matched_count == 6:
        return "Jackpot"
    elif matched_count == 5 and winning_numbers[-1] in user_numbers:
        return "Second prize"
    elif matched_count == 5:
        return "Third prize"
    elif matched_count == 4 and winning_numbers[-1] in user_numbers:
        return "Fourth prize"
    elif matched_count == 4:
        return "Fifth prize"
    elif matched_count == 3 and winning_numbers[-1] in user_numbers:
        return "Sixth prize"
    elif matched_count == 2 and winning_numbers[-1] in user_numbers:
        return "Seventh prize"
    elif matched_count == 3:
        return "Eighth prize"
    else:
        return "NO"

if __name__ == "__main__":
    period = input("enter the drawing date number").replace(' ', '')
    user_numbers_input = input("enter the lottery number").replace(' ', '').split(",")
    winning_numbers = get_lotto_numbers(period)
    
    if winning_numbers is not None:
        result = check_prize([int(i) for i in user_numbers_input], winning_numbers)
        print(f"{result}")
    else:
        print("can't find the drawing date number")
```
