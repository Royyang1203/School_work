以下檔案顯示美國各年齡層男性和女性的百分比，請使用 Matplotlib 將資料視覺化：

1. 使用 subplot 函數將兩個圖表形成一個 2*1 的總圖，圖形大小為 (16,16)。

2. 第一張子圖顯示各年齡男性與女性之長條圖，橫軸單位為年齡，縱軸單位人口(%)，子標題為 The bar chart of population。圖例應該包括「男性」和「女性」。

3. 第二張子圖顯示各男性與女性人口的比例，請以圓餅圖表示，需在圖上顯示人口占比，子標題為 The pie chart of gender。圖例應該包括「男性」和「女性」。

4. 對於軸上的文字，不限制字型、顏色或比例，只要能清晰地閱讀即可。

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv("US_Population_2022.csv")
fig = plt.figure(figsize = (16, 16))  

width = 0.3
index = np.arange(len(df))
ax1 = fig.add_subplot(211)
ax1.bar(index - width/2, df.iloc[:, 1], width, label='Male')
ax1.bar(index + width/2, df.iloc[:, 2], width, label='Female')
plt.xticks(rotation = 60)
plt.xlabel("age")
plt.ylabel(r"percentage of population")
plt.title("The bar chart of population")
plt.legend()

ax2 = fig.add_subplot(212)
plt.pie([df[r'Male % of Population'].sum(), df[r'Female % of Population'].sum()], labels=['Male', 'Female'], autopct = "%1.1f%%")
plt.title("The pie chart of gender")
plt.legend()

fig.savefig("output.png")
```
