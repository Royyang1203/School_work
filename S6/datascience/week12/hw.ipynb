```python
import matplotlib.pyplot as plt
import pandas as pd

url = "https://mlhr.miaoli.gov.tw/tables2.php?y=111&m=4&unit="
table = pd.read_html(url)
```

```python
df = table[0].iloc[:-1, :]  

# print(df)

# fig = plt.figure(figsize = (10, 8), facecolor = "lightblue")  
# fig.suptitle('苗栗縣人口資料', fontsize=16)
# plt.tight_layout()
```

```python
df
```

```python
# 修改列名為英文
column_translation = {
    "鄉鎮別": "Township",
    "鄰數": "Number_of_Neighbors",
    "戶數": "Number_of_Households",
    "男": "Male",
    "女": "Female",
    "合計(男+女)": "Total(Male+Female)",
    "遷入": "Immigration",
    "遷出": "Emigration",
    "出生": "Birth",
    "死亡": "Death",
    "結婚": "Marriage",
    "離婚": "Divorce"
}
df.rename(columns=column_translation, inplace=True)
```

```python
# df.drop("Number_of_Neighbors")
df
```

```python
# 創建鄉鎮別中文到英文的映射
township_translation = {
    "苗栗市": "Miaoli City",
    "苑裡鎮": "Yuanli Town",
    "通霄鎮": "Tongxiao Town",
    "竹南鎮": "Zhunan Town",
    "頭份市": "Toufen City",
    "後龍鎮": "Houlong Town",
    "卓蘭鎮": "Zhuolan Town",
    "大湖鄉": "Dahu Township",
    "公館鄉": "Gongguan Township",
    "銅鑼鄉": "Tongluo Township",
    "南庄鄉": "Nanzhuang Township",
    "頭屋鄉": "Touwu Township",
    "三義鄉": "Sanyi Township",
    "西湖鄉": "Xihu Township",
    "造橋鄉": "Zaoqiao Township",
    "三灣鄉": "Sanwan Township",
    "獅潭鄉": "Shitan Township",
    "泰安鄉": "Tai'an Township",
}
```

```python
df["Township"] = df["Township"].map(township_translation)
```

```python
df.drop('Number_of_Neighbors', axis=1, inplace=True)
```

```python
df
```

```python
df.to_csv("data.csv", index=False)
```

```python
# plt.rcParams['font.sans-serif']=['Microsoft JhengHei'] # 自訂字體
# plt.rcParams['axes.unicode_minus']=False # 為了正常顯示正負號

fig = plt.figure(figsize = (16, 16))  


ax1 = fig.add_subplot(221)
ax1.plot(df.iloc[:, 0], df.iloc[:, 2], label = df.columns[2], linestyle='-')
ax1.plot(df.iloc[:, 0], df.iloc[:, 3], label = df.columns[3], linestyle='-.')
ax1.plot(df.iloc[:, 0], df.iloc[:, 4], label = df.columns[4], linestyle='--')
plt.xticks(rotation = 60)
plt.xlabel("Township")
plt.ylabel("number of people")
plt.title("the line graphs of population")
plt.legend()

ax2 = fig.add_subplot(222)
ax2.plot(df.iloc[:, 0], df.iloc[:, 5], label = df.columns[5], linestyle='-')
ax2.plot(df.iloc[:, 0], df.iloc[:, 6], label = df.columns[6], linestyle='-.')
plt.xticks(rotation = 60)
plt.xlabel("Township")
plt.ylabel("number of people")
plt.title(" the line graphs of immigration and emigration")
plt.legend()

ax3 = fig.add_subplot(223)
ax3.bar(df.iloc[:, 0], df.iloc[:, 1])
plt.xticks(rotation = 60)
plt.xlabel("Township")
plt.ylabel("number of people")
plt.title("the long bar graphs of household number")

ax4 = fig.add_subplot(224)
ax4.plot(df.iloc[:, 0], df.iloc[:, 7], label = df.columns[7], linestyle='-')
ax4.plot(df.iloc[:, 0], df.iloc[:, 8], label = df.columns[8], linestyle='-.')
plt.xticks(rotation = 60)
plt.xlabel("Township")
plt.ylabel("number of people")
plt.title("the line graphs of births and deaths")
plt.legend()


plt.tight_layout()
fig.savefig("week12_hw2.png")
df
```

```python
# %% [markdown]
# # Matplotlib Assignment

# %%
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


# %%
df= pd.read_csv('data.csv')

df.head(len(df))

# %%
plt.figure(figsize=(16, 16), layout="constrained").suptitle(
    "Miaoli County Population Data")

plt.subplot(221)
plt.xlabel("Township")
plt.ylabel("Population Of Each District")
plt.title("Line Graph Showing Total Population")

plt.plot(df["Township"], df["Total(Male+Female)"],
         color='g', linestyle='--', label='Total Population')
plt.plot(df["Male"], color='dodgerblue', label='Male')
plt.plot(df["Female"], color='orange', linestyle='--', label='Female')
plt.legend()

plt.xticks(rotation=90, fontsize=8)

plt.subplot(222)
plt.title("Line Graph Showing the immigration and emigration of each Township")
plt.xlabel("Township")
plt.ylabel("Number of People")

plt.plot(df["Township"], df["Immigration"], color="green", label="Immigration")
plt.plot(df["Township"], df["Emigration"], color="red",
         linestyle="-.", label="Emigration")
plt.xticks(rotation=75, fontsize=8)
plt.legend()

plt.subplot(223)
plt.title("Bar Graph Showing the Number of Housholds for each Township")
plt.xlabel("Township")
plt.ylabel("Number of People")

wrapped_township = df["Township"].to_list()

wrapped_township = [town.replace(' ', '\n') for town in wrapped_township]

plt.grid(axis='y', linewidth=0.2)
plt.xticks(rotation=90)

new_space = np.arange(0, len(df)*2, 2)

plt.bar(new_space, df["Number_of_Households"])

plt.xticks(ticks=new_space, rotation=90, labels=wrapped_township, fontsize=8)

plt.subplot(224)

plt.title("Line Graph Showing the Births and Deaths of each Township")
plt.xlabel("Township", fontsize=10)
plt.ylabel("Number of People", fontsize=10)

wrapped_township = [town.replace(' ', '\n') for town in wrapped_township]

plt.plot(df["Township"], df["Birth"], color="salmon", label="Births")
plt.plot(df["Township"], df["Death"], color="slateblue",
         linestyle="--", label="Deaths")
plt.xticks(rotation=-90, fontsize=8)
plt.grid()
plt.legend()
```
