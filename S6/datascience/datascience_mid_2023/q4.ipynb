Lottery

Write a program that allows users to enter a lottery drawing date number and their lottery numbers, and automatically determines whether they have won a prize. If they have won, output the name of the prize; otherwise, output "NO."

The URL of a lottery website is: "https://www.taiwanlottery.com.tw/lotto/lotto649/history.aspx"

<img src="img/lotto.png">

Input:

The drawing date number consists of nine digits, and the lottery numbers consist of six digits ranging from 1 to 49, with no repeats.

Output:

Winning method: Prize name

Matching all six winning numbers: Jackpot

Matching any five of the winning numbers plus the special number: Second prize

Matching any five of the winning numbers: Third prize

Matching any four of the winning numbers plus the special number: Fourth prize

Matching any four of the winning numbers: Fifth prize

Matching any three of the winning numbers plus the special number: Sixth prize

Matching any two of the winning numbers plus the special number: Seventh prize

Matching any three of the winning numbers: Eighth prize

If the user wins a prize, output the name of the prize. Otherwise, output "NO"


sample input:

112000035

3, 11, 20, 30, 45, 28

sample output:

Second prize

請寫一個程式來讓使用者可以藉由輸入一個期別號以及投注號碼，來自動判斷是否有中獎，如果有請輸出獎項名，沒有則輸出NO。

下列為一樂透網站之網址 : 
"https://www.taiwanlottery.com.tw/lotto/lotto649/history.aspx"

<img src="img/lotto.png">

Input:

期別部分含有九個數字，投注號碼共有六碼，範圍從1~49，不重複

Output:

中獎方式 : 獎項
與當期六個獎號完全相同者 : 頭獎
對中當期獎號之任五碼＋特別號 : 貳獎
對中當期獎號之任五碼 : 參獎
對中當期獎號之任四碼＋特別號 : 肆獎
對中當期獎號之任四碼 : 伍獎
對中當期獎號之任三碼＋特別號 : 陸獎
對中當期獎號之任兩碼＋特別號 : 柒獎
對中當期獎號之任三碼 : 普獎

如果有中獎，就輸出獎項名稱，反之則輸出 NO

sample input:

112000035

3, 11, 20, 30, 45, 28

sample output:

Second prize

```python
import requests
from bs4 import BeautifulSoup

def get_lotto_numbers(period):
    url = f"https://www.taiwanlottery.com.tw/lotto/lotto649/history.aspx"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")

    # Find the target period row
    all_table = soup.find(id='Lotto649Control_history_dlQuery')
    table = all_table.find_all('table')
    for row in table:
        cur_period = row.find_all('td')
        if int(cur_period[5].text) == int(period):
            numbers = [int(number.text) for number in cur_period[13:20]]
            return numbers
    return None

def check_prize(user_numbers, winning_numbers):
    matched_numbers = set(user_numbers).intersection(set(winning_numbers[:-1]))
    matched_count = len(matched_numbers)

    if matched_count == 6:
        return "Jackpot"
    elif matched_count == 5 and winning_numbers[-1] in user_numbers:
        return "Second prize"
    elif matched_count == 5:
        return "Third prize"
    elif matched_count == 4 and winning_numbers[-1] in user_numbers:
        return "Fourth prize"
    elif matched_count == 4:
        return "Fifth prize"
    elif matched_count == 3 and winning_numbers[-1] in user_numbers:
        return "Sixth prize"
    elif matched_count == 2 and winning_numbers[-1] in user_numbers:
        return "Seventh prize"
    elif matched_count == 3:
        return "Eighth prize"
    else:
        return "NO"

if __name__ == "__main__":
    period = input("enter the drawing date number").replace(' ', '')
    user_numbers_input = input("enter the lottery number").replace(' ', '').split(",")
    winning_numbers = get_lotto_numbers(period)
    
    if winning_numbers is not None:
        result = check_prize([int(i) for i in user_numbers_input], winning_numbers)
        print(f"{result}")
    else:
        print("can't find the drawing date number")
```
