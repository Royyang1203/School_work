Please scrape the article titles, points, and links from the following website and convert them into a dataframe or a 2D list. If an article does not have a point number, display "None".

https://news.ycombinator.com/

中文:

請爬取下列網站中的文章標題，points，還有其連結，並轉為dataframe 型別或二維list，如果該文章目前沒有pointy number，則顯示None

https://news.ycombinator.com/

<img src="img/q6.png">

```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd


response = requests.get("https://news.ycombinator.com/")
soup = BeautifulSoup(response.content, 'html.parser')

title_list = soup.find_all(class_='athing')
score_list = soup.find_all(class_='subtext')

res = []

for i, j in zip(title_list, score_list):           
	# print(i.text, j.text)
	title = i.find(class_='titleline').text
	if j.find(class_='score') != None:
		score = re.findall(r'\d+', j.find(class_='score').text)[0]
	else:
		score = None
	if i.find(class_='titleline').find('a', href=True) != None:
		href = i.find(class_='titleline').find('a', href=True)['href']
	else:
		href = None
	res += [[title, score, href]]


df = pd.DataFrame(res, columns=['title', 'score', 'href'])
df
```
