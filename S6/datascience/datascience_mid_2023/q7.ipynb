請使用迴圈敘述撰寫一程式，讓使用者輸入一個正整數（<100），然後以三角形的方式依序輸出1到N的乘法表。

輸入：

一個正整數（<100）

輸出：

以三角形的方式依序輸出此數的乘法結果。

欄寬為4，並應靠右對齊。

多餘的空間請以'.'填滿。

sample input 1:

5

sample output 1:

...1
...2...4
...3...6...9

...5..10..15..20..25

```python


def main():
    n = int(input("請輸入一個正整數（<100）："))
    
    for i in range(1, n+1):
        for j in range(1, i+1):
            result = i * j
            print(f"{result:.>4}", end="")
        print()
    

if __name__ == "__main__":
    main()
```

```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
url = "https://news.ycombinator.com/"
html = requests.get(url)
html.encoding = "utf-8"
sp = BeautifulSoup(html.text,"html.parser")
ans = []
ans2 = []
ans3 = []
ans4 = []
titles = sp.find_all('span',class_="titleline")
scores = sp.find_all('span',class_="score")
herfs = sp.find_all('span',class_="titleline")
for title in titles:
    ans.append(title.text)
for score in scores:
    ans2.append(score.text)
for herf in herfs:
    cc = str(herf.select('a'))
    ans3.append(cc.split('"')[1])
for i in range(29):
    ans4.append([ans[i],ans2[i],ans3[i]])
df = pd.DataFrame(ans4)
df.index = range(1,30)
df.columns = ["Title","Score","Hours"]
df
```
