```python

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support.ui import WebDriverWait
import time


driverPath = "chromedriver.exe"
s = Service(executable_path=driverPath)

# Initialize Chrome WebDriver
driver = webdriver.Chrome(service=s)

# Load CNN webpage
driver.get("https://www.cwb.gov.tw/V8/E/W/County/County.html?CID=10017")


select = Select(driver.find_element(By.CSS_SELECTOR, "#CID.form-control"))
select = Select(driver.find_element(By.CSS_SELECTOR, "#CID.form-control"))

county = input()
# county = 'all'
if county == 'all':
    for i, op in enumerate(select.options):
        if i == 0:
            continue
        print(op.text)
        select.select_by_index(i)
        # driver.execute_script('location.reload()')
        # driver.refresh
        # WebDriverWait(driver, 1)
        # print(driver.current_url)
        today = driver.find_element(By.XPATH, '/html/body/div[2]/main/div/div[2]/div[2]/div[2]/ul/li[1]/span[2]/span[1]')
        # data = driver.find_element(By.CSS_SELECTOR, 'ul.to-to')
        # print(data.text)
        print(today.text)
else:
    select.select_by_visible_text(county)
    today = driver.find_element(By.XPATH, '/html/body/div[2]/main/div/div[2]/div[2]/div[2]/ul/li[1]/span[2]/span[1]')
    print(today.text)


# Close the WebDriver
driver.quit() 
```
