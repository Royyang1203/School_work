## 1.2 K-means

### Definition

- It is an iterative algorithm that divides the unlabeled dataset into k different clusters in such a way that each dataset belongs only one group that has similar properties.

### Meaning

- an unsupervised learning algorithm
- used to solve the clustering problems in machine learning or data science.

### Functions

- Groups the unlabeled dataset into different clusters. 

- K defines the number of pre-defined clusters
     - K=2, there will be two clusters, and for K=3, there will be three clusters, and so on.
     
### Tasks

- Determines the best value for K center points or centroids by an iterative process.
- Assigns each data point to its closest k-center. Those data points which are near to the particular k-center, create a cluster.
     
     
- https://www.javatpoint.com/k-means-clustering-algorithm-in-machine-learning

### Steps

- Step-1: Select the number K to decide the number of clusters.

- Step-2: Select random K points or centroids. (It can be other from the input dataset).

- Step-3: Assign each data point to their closest centroid, which will form the predefined K clusters.

- Step-4: Calculate the variance and place a new centroid of each cluster.

- Step-5: Repeat the third steps, which means reassign each datapoint to the new closest centroid of each cluster.

- Step-6: If any reassignment occurs, then go to step-4 else go to FINISH.

- Step-7: The model is ready.

### Example

```python
# Example

from sklearn.cluster import KMeans
from sklearn import datasets
iris = datasets.load_iris() 
X = iris.data[:, [2,3]] 

kmeansModel = KMeans(n_clusters=3, random_state=46)
clusters_pred = kmeansModel.fit_predict(X)
print('clusters for X:\n', clusters_pred)
print('\ncluster centers:\n', kmeansModel.cluster_centers_)
```

### Parameters for KMeans

- n_cluster
  - int, default=8
  - k value

- random_state
  - int, default=None
  - Determines random number generation for centroid initialization. Use an int to make the randomness deterministic. 
- n_init
  - int, default=10
  - Number of time the k-means algorithm will be run with different centroid seeds. The final results will be the best output of n_init consecutive runs in terms of inertia.
- max_iter
  - int, default=300
  - Maximum number of iterations of the k-means algorithm for a single run.



- https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html

### Attributes for KMeans

- inertia_: inertia_
  - float
  - Sum of squared distances of samples to their closest cluster center
- cluster_centers_： 
  - ndarray of shape (n_clusters, n_features)
  - Coordinates of cluster centers.
- labels_
  - ndarray of shape (n_samples,)
  - Labels of each point

### Methods for KMeans

- fit(X[, y, sample_weight])
  - Train k-means clustering
- predict(X[, sample_weight])
  - Predict the closest cluster each sample in X belongs to.
- fit_predict(X[, y, sample_weight])
  - Call fit( ) and then call predict( )
  - Compute cluster centers and predict cluster index for each sample.
- transform(X)
  - Transform X to a cluster-distance space.
- fit_transform(X[, y, sample_weight])
  - Call fit( ) and then call transform( )
  - Compute clustering and transform X to cluster-distance space.



- https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html


```python
import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn import cluster
import matplotlib.pyplot as plt

iris = datasets.load_iris()

X = pd.DataFrame(iris.data[:, [2,3]])
X.columns = ["petal_length","petal_width"]
y = iris.target
```

```python
k = 3

kmeans = cluster.KMeans(n_clusters=k, random_state=12)
kmeans.fit(X)
print("kmeans.labels_:\n", kmeans.labels_)
print("\ncluster centers:\n", kmeans.cluster_centers_)

new_X = pd.DataFrame([[3,4], [1,3], [2,4]], columns = ["petal_length","petal_width"])

clusters_pred = kmeans.predict(new_X)
print('\nPrecdicted y:', clusters_pred)
```

```python
# Distance between samples and every cluster center
import math

distances = kmeans.transform(X)
print("\nTransform X to a cluster-distance space:\n", distances[:5])
print('\nthe fist 5 samples:\n', X.head(5))
print('\ncluster center:\n', kmeans.cluster_centers_)   # calculate distance by kmeans
print()

for i in range(0, 3):     # verify distance by formula
  print('\ndistance between sample 1', 'with cluster center', i, '\n')
  distance = math.sqrt((X['petal_length'][0]-kmeans.cluster_centers_[i,0])**2+(X['petal_width'][0]-kmeans.cluster_centers_[i,1])**2)
  print(distance)
```

## 3.2 Finding the optimal k

- The performance of the K-means clustering algorithm depends upon highly efficient clusters that it forms. 
- But choosing the optimal number of clusters is a big task. 


```python
# the number of k affects the results

import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn import cluster
import matplotlib.pyplot as plt

iris = datasets.load_iris()

X = pd.DataFrame(iris.data[:, [2,3]])
X.columns = ["petal_length","petal_width"]
y = iris.target
k = 2

print(X)
print(X["petal_length"])

kmeans = cluster.KMeans(n_clusters=k, random_state=12)
kmeans.fit(X)

colmap = np.array(["r", "y"])

print(colmap)
print(kmeans.labels_)
plt.figure(figsize=(10,5))
plt.subplot(1, 2, 1)
plt.scatter(X["petal_length"], X["petal_width"], 
            color=colmap[kmeans.labels_])

plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.title("K-means Clustering (k=2)")

plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1], marker='X', s=100)


k = 3

kmeans = cluster.KMeans(n_clusters=k, random_state=12)
kmeans.fit(X)

plt.subplot(1, 2, 2)
colmap = np.array(["r", "b", "y"])
# plt.figure(figsize=(10,5))

plt.scatter(X["petal_length"], X["petal_width"], 
            color=colmap[kmeans.labels_])

plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.title("K-means Clustering (k=3)")

plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1], marker='X', s=100)

plt.show()

plt.show()
```

 ### Elbow method

 - A popur approach of finding the optimal number of clusters.
 - This method uses the concept of WCSS value. 
  - WCSS stands for Within Cluster Sum of Squares, which defines the total variations within a cluster. 
 - The formula
  - https://www.javatpoint.com/k-means-clustering-algorithm-in-machine-learning

### steps of Elbow method:

- It executes the K-means clustering on a given dataset for different K values (ranges from 1-10).
- For each value of K, calculates the WCSS value.
- Plots a curve between calculated WCSS values and the number of clusters K.
- The sharp point of bend or a point of the plot looks like an arm, then that point is (elbow point) considered as the best value of K.


#### Since the graph shows the sharp bend, which looks like an elbow, hence it is known as the elbow method.


- https://www.javatpoint.com/k-means-clustering-algorithm-in-machine-learning

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import datasets
iris = datasets.load_iris() 
X = iris.data[:, [2,3]] 
```

```python
plt.figure(figsize=(10,5))
wcss_list= [] 
for i in range(1, 11):  
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state= 42)  
    kmeans.fit(X)  
    wcss_list.append(kmeans.inertia_)  
plt.plot(range(1, 11), wcss_list)  
plt.title('The Elobw Method Graph')  
plt.xlabel('Number of clusters(k)')  
plt.ylabel('wcss_list')  
plt.show()  

# the optimal number of K is 3 
```

### Accuracy of clustring by confusion matrix <a name = "Accuracy">
   

```python
# accuracy of clustring by confusion matrix

import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn import cluster
import sklearn.metrics as sm

iris = datasets.load_iris()

# X = pd.DataFrame(iris.data, columns=iris.feature_names)
# X.columns = ["sepal_length","sepal_width","petal_length","petal_width"]
# y = iris.target
# k = 3

X = pd.DataFrame(iris.data[:, [2,3]])
X.columns = ["petal_length","petal_width"]
y = iris.target
k = 3

kmeans = cluster.KMeans(n_clusters=k, random_state=12)
kmeans.fit(X)

pred_y = np.choose(kmeans.labels_, [2, 0, 1]).astype(np.int64) # adjust above clusters 

print('Accuracy:', sm.accuracy_score(y, pred_y))

print('\nConfusion matrix:\n', sm.confusion_matrix(y, pred_y))
```

### Advantages

- It is very easy to understand and implement.

- If we have large number of variables then, K-means would be faster than Hierarchical clustering.

- On re-computation of centroids, an instance can change the cluster.

- Tighter clusters are formed with K-means as compared to Hierarchical clustering.

### Disadvantages

- It is a bit difficult to predict the number of clusters i.e. the value of k.

- Output is strongly impacted by initial inputs like number of clusters (value of k).

- Order of data will have strong impact on the final output.

- It is very sensitive to rescaling. If we will rescale our data by means of normalization or standardization, then the output will completely change.final output.

- It is not good in doing clustering job if the clusters have a complicated geometric shape.
