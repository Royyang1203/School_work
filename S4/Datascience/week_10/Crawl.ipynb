# Introduction

## 何為爬蟲?

* 「網頁爬蟲」(web Scraping, web crawl)或「網頁資料擷取」(web data extraction)
* 藉由程式模擬瀏覽器通訊過程，取得特定網站上之特定資料
* 剖析文件，取出所需部分，整理資料
### 步驟
* 確認要抓取的網站及內容
    * 藉由瀏覽器之開發人員工具找到HTML文件中我們所要之資料的詳細資訊
        * 找到位置以及資料的特徵，例如:標籤名稱和屬性值。
    * 爬取網頁
        * 靜態網頁 : 使用HTTP函式庫
            * 與web伺服器進行HTTP通訊的函數庫，以便取得回應文件的HTML網頁內容，可以使用Requests套件
        * 動態網頁
            * 使用 selenium
    * 剖析網頁
        * 網頁爬工具或函數庫
            * 在取得回應文件後，需要工具或函束褲來剖析文件，以便取出所需的資料，可以使用BeautifulSoup套件。
     * 資料處理或存檔
         * numpy
         * pandas
         * matplotlab

```python
# 爬蟲範例

import requests
url = 'https://www.ndhu.edu.tw'
html = requests.get(url)
html.encoding="utf-8"
print(html.text)
```

```python
import requests

url = 'https://www.ndhu.edu.tw'
html = requests.get(url)
html.encoding="utf-8"
if html.status_code == requests.codes.ok: # 疰ฎ 200
    print(html.url)
    print(html.status_code)
    print(requests.codes.ok)
    print('file size:', len(html.text))
```

```python
# 連線失敗

import requests

url = 'http://aaa.24ht.com.tw/'
html = requests.get(url)
print(html.status_code)
print(html.raise_for_status) # 連線狀態碼
```

```python
# get headers of websites' response

import requests

r = requests.get("http://www.google.com")
print(r.headers['Content-Type'])
print(r.headers['Content-Length'])
print(r.headers['Date'])
print(r.headers['Server'])
```
