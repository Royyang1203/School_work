```python
import matplotlib.pyplot as plt
import pandas as pd

plt.rcParams['font.sans-serif']=['Microsoft JhengHei'] # 自訂字體
plt.rcParams['axes.unicode_minus']=False # 為了正常顯示正負號

url = "https://mlhr.miaoli.gov.tw/tables2.php?y=111&m=4&unit="
table = pd.read_html(url)


df = table[0].iloc[:-1, :]  

# print(df)

fig = plt.figure(figsize = (10, 8), facecolor = "lightblue")  
fig.suptitle('苗栗縣人口資料', fontsize=16)


ax1 = fig.add_subplot(221)
ax1.plot(df.iloc[:, 0], df.iloc[:, 3], label = df.columns[3], linestyle='-')
ax1.plot(df.iloc[:, 0], df.iloc[:, 4], label = df.columns[4], linestyle='-.')
ax1.plot(df.iloc[:, 0], df.iloc[:, 5], label = df.columns[5], linestyle='--')
plt.xticks(rotation = 60)
plt.xlabel("行政區域")
plt.ylabel("人數")
plt.title("男、女人口及合計之折線圖")
plt.legend()

ax2 = fig.add_subplot(222)
ax2.plot(df.iloc[:, 0], df.iloc[:, 6], label = df.columns[6], linestyle='-')
ax2.plot(df.iloc[:, 0], df.iloc[:, 7], label = df.columns[7], linestyle='-.')
plt.xticks(rotation = 60)
plt.xlabel("行政區域")
plt.ylabel("人數")
plt.title("遷入遷出之折線圖")
plt.legend()

ax3 = fig.add_subplot(223)
ax3.bar(df.iloc[:, 0], df.iloc[:, 2])
plt.xticks(rotation = 60)
plt.xlabel("行政區域")
plt.ylabel("人數")
plt.title("戶數之長條圖")

ax4 = fig.add_subplot(224)
ax4.plot(df.iloc[:, 0], df.iloc[:, 8], label = df.columns[8], linestyle='-')
ax4.plot(df.iloc[:, 0], df.iloc[:, 9], label = df.columns[9], linestyle='-.')
plt.xticks(rotation = 60)
plt.xlabel("行政區域")
plt.ylabel("人數")
plt.title("出生、死亡之折線圖")
plt.legend()


plt.tight_layout()
fig.savefig("week12_hw2.png")
df
```
