```python
import matplotlib.pyplot as plt
import numpy as np

a = int(input("please enter a ->"))
b = int(input("please enter b ->"))
c = int(input("please enter c ->"))

x = np.arange(-50, 50)
y = [(a * i * i + b * i + c) for i in x]
plt.plot(x, y, label = "Y = {} * X^2 {:+} * X {:+}".format(a, b, c))
plt.title("Y = {} * X^2 {:+} * X {:+}".format(a, b, c))  # add title
plt.legend()  #放置圖例
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv(r"C:\Coding\43_School_S3\VIP\week_6\2002.csv",
                 index_col=0, parse_dates=True)

pic = plt.figure(figsize=(8, 4))  # 設定圖形大小

close = pic.add_axes([0, 0.2, 1, 1])
close.plot(df.index, df.Close)

high = df.High.max()
low = df.Low.min()
close.axhline(y=high, xmin=0, xmax=1, color="Red", label=high)
close.axhline(y=low, xmin=0, xmax=1, color="Green", label=low)

plt.title("2002")
plt.ylabel("price")
close.legend()

volume = pic.add_axes([0, 0, 1, 0.2])
volume.bar(df.index, df['Volume'], color='black')
plt.ylabel("volume")
plt.xlabel("day")
      
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df_2892 = pd.read_csv(r"2892_202110.csv", parse_dates=True)
df_5880 = pd.read_csv(r"5880_202110.csv", parse_dates=True)

fig = plt.figure(figsize=(15, 10))

plt.plot(df_2892["day"], df_2892["price"], label=2892)
plt.plot(df_5880["day"], df_5880["price"], label=5880)

plt.title("2892 v.s. 5880")
plt.xlabel("time")
plt.ylabel("price")

plt.gcf().autofmt_xdate()
plt.grid()
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df_2330 = pd.read_csv(r"C:\Coding\43_School_S3\investment\week06\2330.csv", index_col=0, parse_dates=True)

df = df_2330.loc["2019-01-01":"2020-12-31", : ]
pic = plt.figure(figsize=(15, 15))

p1 = plt.subplot(221)
p1.plot(df.Close)

plt.title("Close")
plt.xlabel("time")
plt.ylabel("price")

p2 = plt.subplot(222)
p2.plot(df.High, label="a")

plt.title("High")
plt.xlabel("time")
plt.ylabel("price")
plt.legend()

p1 = plt.subplot(223)
p1.plot(df.Low)

plt.title("Low")
plt.xlabel("time")
plt.ylabel("price")

p1 = plt.subplot(224)
p1.bar(df.index, df["Volume"], color="red")

plt.title("Volume")
plt.xlabel("time")
plt.ylabel("number")

plt.show()

```
