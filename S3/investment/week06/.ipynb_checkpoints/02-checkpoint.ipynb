```python
import mplfinance as mpf
import matplotlib.pyplot as plt
import pandas as pd

_df = pd.read_csv(
    r"C:\Coding\43_School_S3\VIP\week_6\2002.csv",
    index_col=0, parse_dates=True)

cut = _df.loc["2019-01":"2021-09", :]
# mpf.figure(num=1, figsize=(15, 18))
# 設置k線圖顏色
my_color = mpf.make_marketcolors(up='red',  # 上漲
                                 down='green',  # 下跌
                                 edge='i',  # 隱藏K線邊緣
                                 volume='in',  # 成交量用同樣的顏色
                                 inherit=True)

my_style = mpf.make_mpf_style(gridaxis='both',  # 設置網格
                              gridstyle='-.',
                              y_on_right=True,
                              marketcolors=my_color)

mpf.plot(cut, type="candle", mav=(5, 10, 20, 60),
         volume=True, style=my_style, show_nontrading=False,
         figratio=(18, 10), figscale=1.2, tight_layout=True)
```
