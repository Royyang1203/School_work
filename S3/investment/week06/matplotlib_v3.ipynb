# 折線圖

使用前需要先匯入模組：
   import matplotlib.pyplot as plt
    
plot() 函式在座標系統繪製線條或標記，其語法如下：
plot(*args[, 選擇性參數1 = 值1, 選擇性參數2 = 值2, …])
   *args:為不限定個數的參數，可以包含多對的x和y座標的串列(list)，以及選擇性的格式化字串
     ，格式化字串包含線條的顏色、大小、形狀等
    例如：
        x = [1, 3, 5]
        y = [0, 1, 4]
        plt.plot(x, y)    
        plt.plot(x)   
        plt.plot(x, y, "go")

設定完圖形後，要使用plt.show()將圖形顯示出來

```python
import matplotlib.pyplot as plt
x = [1, 3, 5]
y = [0, 1, 2]


plt.plot(x)  
  # 在圖上依序畫出 1, 3, 5的值，因此這些點為：(0,1),(1,3),(2,5)
plt.show()

plt.plot(x, y)  
  #畫出x和y串列所形成的座標的點的連線，這些點包含(1,0), (3,1), (5,2)
plt.show()


plt.plot(x, y, 'go')
   #第三個參數 'g'表示green, 'o'表示以點的方式呈現
plt.show()
```

optional properties 
refer to http://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot

matplotlib.pyplot 的 plot 繪圖方法，除了 x 坐標串列及 y 坐標串列為必要參數外，還有數十個選擇性參數設定繪圖特性，下面是常用的 4 個選擇性參數：

```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(0, 10, 50)
sinus = np.sin(x)
cosinus = np.cos(x)
plt.plot(x, sinus, "r-o", label="sin(x)")
plt.plot(x, cosinus, "g--", label="cos(x)")
plt.legend()    #圖例位置以預設方式呈現
plt.xlabel("Rads")
plt.ylabel("Amplitude")
plt.title("Sin and Cos Waves")
plt.show()
```

### 使用subplot( )繪製多張圖表

* 使用matplotlib.pyplot模組的subplot() 函式在視窗內繪製多張圖表，其語法如下：  
    * subplot(nrows, ncols, plot_number)  
        * subplot(2, 3, 1) # 共有2列的子圖表，共有3行,要在第 1張圖繪圖 (總共會有2*3=6張圖，由左到右，上到下排列，也就是先水平再垂直） 
        * 可簡寫為： subplot(231)  <br /> <br /> 

```python
# 14-10
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10,10,100)
y1 = 20 * np.sin(x)
y2 = x * x * np.cos(x) + 0.5

plt.subplot(211)
plt.plot(x, y1, "b-")

plt.subplot(212)
plt.plot(x, y2, "r--")
plt.show()
```

### 使用 add_axes( )繪製多張圖表

* 格式：add_axes([左下角x座標, 左下角y座標, 右上角x座標, 右上角y座標])
* 座標的值為 0 ~ 1 ，原點(0,0)的位置在左下角
    * fig = plt.figure(參數)
    * pic1 = fig.add_axes([0,0.3,1,0.5])   
    * pic1.plot(參數)

```python
# 使用 add_axes( )繪製多張圖表
import matplotlib.pyplot as plt

#建立初始圖框
fig = plt.figure(figsize=(20, 15))

#在fig圖裡可以堆疊，自訂放置位置
ax = fig.add_axes([0,0.3,1,0.5])   #  最上方的圖
ax2 = fig.add_axes([0,0.1,1,0.2])  #  中間的圖
ax3 = fig.add_axes([0,0,1,0.1])    #  最下方的圖
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10,10,100)

#建立初始圖框
fig = plt.figure(figsize=(20, 15))

#在fig圖裡可以堆疊，自訂放置位置
ax = fig.add_axes([0,0.3,1,0.5])   #  最上方的圖
ax2 = fig.add_axes([0,0.1,1,0.2])  #  中間的圖
ax3 = fig.add_axes([0,0,1,0.1])    #  最下方的圖
ax1.plot(x, np.power(x, 1))
ax2.plot(x, np.power(x, 2))
ax3.plot(x, np.power(x, 3))

plt.show()
```

## 繪製具時間的圖表（例如股票日線圖）

* 可使用plt.plot_date() 取代 plt.plot()，使用方式一樣，但若是x或y 座標為日期格式時，自動顯示時間格式
* 目前plt.plot()也有此功能了

```python
import datetime
import matplotlib.pyplot as plt

today = datetime.date.today()
step = datetime.timedelta(2)

x = []
y = []
for i in range(10):
   x.append(today+step*i) 
   y.append(i+2)  
plt.figure(figsize = (15,8)) # 設定圖形大小
plt.plot_date(x, y, 'g-')
plt.show()
```

# 長條圖（Bar Plots）

* 是使用長條型色彩區塊的高和長度來顯示分類資料，我們可以顯示水平或垂直方向的長條圖
* 使用matplotlib.pyplot模組的bar() 函式繪製長條圖
* bar(X, Y, width=5, label='str', tick_label=lst)


```python
import matplotlib.pyplot as plt

x = [70, 80, 90, 100, 110, 120, 130, 140, 150]
y = [2.2, 5.3, 11.5, 19.7, 22.9, 19.6, 11.2, 5.5, 2.1]
tl = ["<75", "75~84", "85~94", "95~104", "105~114", "115~124", "125~134", "135~144", ">144"]

plt.figure(figsize = (8, 4))  # figure size = 8-inch width, 4-inch heigh
# plt.bar(left = x, height = y, width = 5, label = "Sample1", tick_label = tl)
plt.bar(x, y, width = 5, label = "Sample1", tick_label = tl)
plt.legend()
plt.xlabel("Smarts")
plt.ylabel("Probability (%)")
plt.title("Bar of IQ")
plt.show()
```

# 直方圖

* 用來顯示數值資料的分佈，屬於一種次數分配表，
* 可以使用長方形面積來顯示變數出現的頻率，寬度是分割區間。
* 在實務上，我們可以使用直方圖來評估連續變數的機率分佈

* 使用matplotlib.pyplot模組的hist() 函式繪製直方圖，其語法如下：
 * hist(x[, [bin個數, ...])

# 圓餅圖

* 派圖（Pie Plots）也稱為圓餅圖（Circle Plots）
 * 這是使用一個完整圓形來表示統計資料的圖表，如同是在切圓形蛋榚，以不同切片大小來標示資料的比例

* matplotlib.pyplot模組的pie() 函式繪製圓形圖

* 語法： 
 * pie(x[, 選擇性參數1 = 值1, 選擇性參數2 = 值2, …])

* 「資料串列」是一個數值串列，為畫圓餅圖的資料，其為必要參數。
* 「選擇性參數串列」可有可無，參數名稱及功能為：  
 * labels：每一個項目標題組成的串列。
 * colors：每一個項目顏色組成的串列。
 * explode：每一個項目凸出數值組成的串列，「0」表示正常顯示未爆出
 * labeldistance：項目標題與圓心的距離是半徑的多少倍，例如「1.1」表示項目標題與圓心的距離是半徑的 1.1 倍。
 * autopct：項目百分比的格式，語法為「% 格式 %%」，例如「%2.1f%%」表示整數 2 位數，小數 1 位數。
 * shadow：布林值，True 表示圖形有陰影，False 表示圖形沒有陰影。
 * startangle：開始繪圖的起始角度，繪圖會以逆時針旋轉計算角度。
 * pctdistance：百分比文字與圓心的距離是半徑的多少倍。

* 預設繪製的圓餅圖是橢圓形，若要繪製正圓形圓餅圖，需以下列方法讓 x、y 軸單位相等：

```python
# 陳允傑，Python 資料科學與人工智慧 Ch10_2_4

labels = ["Python", "C++", "Java", "JS", "C", "C#"]
ratings = [5, 6, 15, 3, 12, 4]
 
plt.pie(ratings, labels=labels)
plt.title("Programming Language Usage") 
plt.axis("equal")
plt.show()
```

```python
# 派圖可以使用突增值的元組或清單來標示切片是否需要突出來強調顯示
# 陳允傑，Python 資料科學與人工智慧 Ch10_2_4

labels = ["Python", "C++", "Java", "JS", "C", "C#"]
ratings = [5, 6, 15, 3, 12, 4]
explode = (0, 0, 0, 0.2, 0, 0.2)
 
plt.pie(ratings, 
        labels=labels,
        explode=explode)
plt.title("Programming Language Usage") 
plt.axis("equal")
plt.show()
```

```python
# 顯示切片色彩的圖例
# 陳允傑，Python 資料科學與人工智慧 Ch10_2_4

labels = ["Python", "C++", "Java", "JS", "C", "C#"]
ratings = [5, 6, 15, 3, 12, 4]
explode = (0, 0, 0, 0.2, 0, 0.2)
 
patches, texts = plt.pie(ratings, 
                         labels=labels,
                         explode=explode)
plt.legend(patches, labels, loc="best")
plt.title("Programming Language Usage") 
plt.axis("equal")
plt.show()
```

```python
import matplotlib.pyplot as plt
activities = ["Math", "Chinese", "Programming", "others"]
hours = [5, 7, 2, 7]
colors = ["lightgreen", "lightblue", "yellow", "pink"]	
plt.pie(hours, labels = activities, colors = colors, shadow = True, explode = (0, 0.1, 0, 0), 
        autopct = "%1.1f%%")
plt.axis("equal")
plt.show()
```

## 從網頁抓取資料後顯示
範例：ch07. personplot 桃園縣人口資料

```python
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup as bs
import requests

year = []
person = []
url = "http://www.daxi-hro.tycg.gov.tw/home.jsp?id=25&parentpath=0,21,22"
content = requests.get(url).text
parse = bs(content, "html.parser")
data1 = parse.select("table[summary^='歷年戶數統計列表排版用']")[0]
rows = data1.find_all("tr")
for row in rows:
    cols = row.find_all("td")
    if(len(cols) > 0):
        if cols[1].text!="─":
            year.append(cols[0].text[:-1])
            person.append(cols[1].text)
plt.plot(year, person, linewidth=2.0)
plt.title("桃園市大溪區歷年戶數")
plt.xlabel("年度")
plt.ylabel("戶數")
plt.show()
```
